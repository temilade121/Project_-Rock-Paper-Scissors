```python
# import Numpy library as np

import numpy as np

 

# input the number of student and number of subject

 

num_of_student = int(input("Enter the number of student in class : "))

num_of_subject = int(input("Enter the number of subject each student offer : "))

 

 

# Creating a numpy array to store the marks of each student in each subject

 

marks = np.zeros((num_of_student, num_of_subject), dtype=int)

 

# input mark for each student and subject

 

for i in range(num_of_student):

    for n in range(num_of_subject):

        marks[i][n]= int(input(f" Enter mark for student{i+1}, subject{n+1} :" ))

 

# Calculating the total marks and percentage for each student

 

total_marks = np.sum(marks, axis = 1)

percentage = (total_marks/(num_of_subject*100)) * 100

 

# Defining a grade by assigning grades based on the percentage

 

def grade(percentage):

    if percentage >= 80:

        return "A"

    elif percentage >=70:

        return "B"

    elif percentage >= 60:

        return "C"

    elif percentage >= 50:

        return "D"

    elif percentage >= 40:

        return "E"

    else:

        return "F"

   

# calculate the grades for each student

 

grades = [grade(p) for p in percentage]

 

# result

print("\nResult Summary:")

print("----------------------------------------")

print("Student\tTotal Marks\tPercentage\tGrade")

print("-----------------------------------------")

 

for i in range(num_of_student):

    print(f" Student{i+1}\t{total_marks[i]}\t\t{percentage[i]: .2f}%\t\t{grades[i]}")

print("-----------------------------------------")

 

 


 

```
